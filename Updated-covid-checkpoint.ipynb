```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier # Import Decision Tree Classifier
from sklearn.model_selection import train_test_split # Import train_test_split function
from sklearn import metrics #Import scikit-learn metrics module for accuracy calculation
```

```python
pima = pd.read_csv("covid.csv")
```

```python
feature_cols = ['Admission_DATE', 'PatientAge', 'Gender','WhatkindofTreatmentprovided','Ventilated','Redbloodcelldistributionwidth', 'Monocytes', 'Whitebloodcellcount', 'PlateletCount','LymphocyteCount','NeutrophilsCount']
X = pima[feature_cols] # Features
y = pima.Outcome # Target variable
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le_Admission_DATE = LabelEncoder()
le_PatientAge = LabelEncoder()
le_Gender = LabelEncoder()
le_WhatkindofTreatmentprovided = LabelEncoder()
le_Ventilated = LabelEncoder()
le_Redbloodcelldistributionwidth = LabelEncoder()
le_Monocytes = LabelEncoder()
le_Whitebloodcellcount = LabelEncoder()
le_PlateletCount = LabelEncoder()
le_LymphocyteCount = LabelEncoder()
le_NeutrophilsCount = LabelEncoder()
```

```python
pima['Admission_DATE'] = le_Admission_DATE.fit_transform(pima['Admission_DATE'])
pima['PatientAge'] = le_PatientAge.fit_transform(pima['PatientAge'])
pima['Gender'] = le_Gender.fit_transform(pima['Gender'])
pima['WhatkindofTreatmentprovided'] = le_WhatkindofTreatmentprovided.fit_transform(pima['WhatkindofTreatmentprovided'])
pima['Ventilated'] = le_Ventilated.fit_transform(pima['Ventilated'])
pima['Redbloodcelldistributionwidth'] = le_Redbloodcelldistributionwidth.fit_transform(pima['Redbloodcelldistributionwidth'])
pima['Monocytes'] = le_Monocytes.fit_transform(pima['Monocytes'])
pima['Whitebloodcellcount'] = le_Whitebloodcellcount.fit_transform(pima['Whitebloodcellcount'])
pima['PlateletCount'] = le_PlateletCount.fit_transform(pima['PlateletCount'])
pima['LymphocyteCount'] = le_LymphocyteCount.fit_transform(pima['LymphocyteCount'])
pima['NeutrophilsCount'] = le_NeutrophilsCount.fit_transform(pima['NeutrophilsCount'])
```

```python
feature_cols = ['Admission_DATE', 'PatientAge', 'Gender','WhatkindofTreatmentprovided','Ventilated','Redbloodcelldistributionwidth', 'Monocytes', 'Whitebloodcellcount', 'PlateletCount','LymphocyteCount','NeutrophilsCount']
X = pima[feature_cols] # Features
y = pima.Outcome # Target variable
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1) # 80% training and 20% test
```

```python
clf = DecisionTreeClassifier()

# Train Decision Tree Classifer
clf = clf.fit(X_train,y_train)

#Predict the response for test dataset
y_pred = clf.predict(X_test)
```

```python
# Model Accuracy, how often is the classifier correct?
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```
